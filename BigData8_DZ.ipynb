В третьем блоке текст - это ссылка на другую новость. Это нам не нужно

```python
soup.find_all(class_ = 'article__block')[2].get_text()
```

Запускаем цикл, где из каждого блока новости выбиратся текст и объединяется в одну строку в случае, если текст не содержится в блоке article__article (блок со ссылкой на другие новости)

```python
text = ''
for block in soup.find_all(class_ = 'article__block'):
  if block.find('div', class_="article__article"):
    pass
  else:
    text = ''.join([text, block.get_text()])
```

```python
text
```

# Удаление стоп-слов

Устанавливаем библиотеку NLTK и загружаем словарь стоп-слов (не имеющих смысла слов)

```python
pip install nltk
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords_ru = stopwords.words("russian")
stopwords_ru[0:11]
```

Разбиваем строку с текстом на слова

Попробуем с помощью метода split()

```python
text.split()
```

С помощью метода split() не разделяются те слова, между которыми находятся точки

Попробуем использовать функцию split() из библиотеки re

```python
import re
```

```python
words = [word.lower() for word in re.split(';|\.| ', text) if word.lower() not in stopwords_ru]
text = ' '.join(words)
```

```python
text
```

# Анализ настроения новости

```python
from dostoevsky.tokenization import RegexTokenizer
from dostoevsky.models import FastTextSocialNetworkModel

tokenizer = RegexTokenizer()
model = FastTextSocialNetworkModel(tokenizer=tokenizer)
data = model.predict([text])
```

```python
data
```
